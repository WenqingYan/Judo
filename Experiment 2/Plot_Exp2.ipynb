```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
import numpy as np
from numpy import NaN
import matplotlib
from pylab import rcParams

#%matplotlib widget, camera ready submission problem
matplotlib.rc('pdf', fonttype=42)
matplotlib.rcParams['savefig.dpi'] = 300
rcParams.update({'figure.autolayout': True})

def readfile(filename):
    types = {'frequency': float,'power': float}
    df = pd.read_csv(filename, skiprows=0, header=None, delim_whitespace=False, delimiter=' ')
    df.columns = ['frequency', 'power']
    df.dropna(inplace=True)
    #print(df.describe())
    return df
```

```python
path = './logs/i2/'
filename = 'data0'
df_1 = readfile(path + filename + ".csv")

path = './logs/i4/'
filename = 'data0'
df_2 = readfile(path + filename + ".csv")

color = ["#77A136", "#EF9136"]

# Visualization
# Plot
from pylab import rcParams
rcParams.update({'figure.autolayout': True})
rcParams['figure.figsize'] = 8, 4
fig, (ax1,ax2) = plt.subplots(2,1)

ax1.plot(df_1['frequency']/1e6, df_1['power'], color=color[1], linewidth=2)
ax1.set_title('Square baseband signal', fontsize=20)
ax1.tick_params(labelbottom=False, labelsize=14)
ax1.grid()

ax2.plot(df_2['frequency']/1e6, df_2['power'], color=color[0], linewidth=2)
ax2.tick_params(labelsize=14)
ax2.set_title('Sinusoidal baseband signal', fontsize=20)
ax2.set_xlabel('Frequency [MHz]', fontsize=20)
ax2.grid()

fig.text(0.0, 0.54, 'Signal strength [dBm]', va='center', ha='center', rotation='vertical', fontsize=20)
plt.tight_layout()
plt.savefig('./square_sin.pdf', bbox_inches = "tight")
```
